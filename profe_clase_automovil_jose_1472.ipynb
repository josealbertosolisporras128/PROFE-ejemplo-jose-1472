```python
class automovil:
    def __init__(self, marca, placas, precio):
        self.marca = marca
        self.placas = placas
        self.precio = precio

    def leerdatos(self):
        self.marca = input("ingrese marca: ")
        self.placas = input("ingrese placas: ")
        # Ensure precio is stored as a string or converted to a number if needed
        self.precio = input("ingrese precio: ")
        # No need to return these, as they are stored in self

    def mostrardatos(self):
        print("marca: ", self.marca)
        print("placas: ", self.placas)
        print("precio: ", self.precio)
        # No need to return these, as the purpose is to print

# fin de la clase automovil


print("carro numero 1")
carro1 = automovil("", "", "") # creando objeto carro 1
carro1.leerdatos()
carro1.mostrardatos()
print("----------------------")
print("carro numero 2")
carro2 = automovil("", "", "") # creando objeto carro 2
carro2.leerdatos()
carro2.mostrardatos()
print("----------------------")
print("carro numero 3")
carro3 = automovil("", "", "") # creando objeto carro 3
carro3.leerdatos()
carro3.mostrardatos()
print("sofware desarrollado por jose solis NL 51")
```
